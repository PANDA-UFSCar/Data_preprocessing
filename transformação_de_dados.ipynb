#Transformação de dados


A transformação de dados é uma etapa fundamental na preparação de dados, que envolve a aplicação de técnicas e métodos para preparar e manipular os dados brutos, tornando-os adequados para análise e modelagem. A importância da transformação de dados reside no fato de que os dados coletados do mundo real muitas vezes apresentam complexidades, irregularidades e peculiaridades que podem dificultar a análise direta ou prejudicar a eficácia dos algoritmos de aprendizado de máquina.

Algumas técnicas:

1.   Tratamento de Valores Ausentes: Muitos conjuntos de dados reais contêm valores ausentes, que podem afetar a qualidade dos resultados de análise e modelagem. A transformação de dados inclui técnicas para tratar esses valores ausentes de maneira adequada, como preenchimento com média, mediana ou outros métodos de imputação.


2.   Padronização e Normalização: Ao transformar os dados para uma escala comum, como a padronização ou normalização, é possível evitar que certos atributos dominem os outros durante a análise e modelagem. Isso é especialmente importante em algoritmos sensíveis à escala, como regressão linear.

3.   Criação de Novas Variáveis (Feature Engineering): A transformação de dados permite a criação de novas variáveis com base nas informações existentes. Essas novas variáveis podem conter informações mais relevantes ou fornecer insights adicionais para análise e modelagem.

##Padronização


É o redimensionamento dos recursos para criar um “padrão” garantindo que os novos dados tenham média zero e desvio padrão igual a 1

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler

```python
# Importando o StandardScaler

from sklearn.preprocessing import StandardScaler
```

```python
# Selecionar apenas as colunas numéricas para padronização
colunas_numericas = dados.select_dtypes(include=[float, int]).columns
# selecionamos apenas as colunas numéricas usando select_dtypes

# Criar um objeto de padronização
scaler = StandardScaler()

# Padronizar os dados
dados[colunas_numericas] = scaler.fit_transform(dados[colunas_numericas])

dados
```

Após a execução, as colunas numéricas do DataFrame estarão padronizadas, ou seja, terão média zero e desvio padrão igual a um, o que é útil para muitos algoritmos de aprendizado de máquina que podem se beneficiar da mesma escala entre os atributos.

##Normalização

Uma alternativa a padronização é normalizar os dados entre um valor mínimo e máximo (geralmente entre 0 e 1)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Selecionar apenas as colunas numéricas para normalização
colunas_numericas = dados.select_dtypes(include=[float, int]).columns

# Criar um objeto de normalização
scaler = MinMaxScaler()

# Normalizar os dados
dados[colunas_numericas] = scaler.fit_transform(dados[colunas_numericas])

dados
```

Após a execução, as colunas numéricas do DataFrame estarão normalizadas, ou seja, seus valores estarão no intervalo [0, 1], com base no mínimo e máximo de cada coluna.

###Normalizando dados dispersos

```python
from sklearn.preprocessing import MaxAbsScaler
```

O MaxAbsScaler funciona de uma maneira semelhante, porém dividindo todos os dados pelo máximo absoluto daquela coluna.

Isso faz com que o range seja de -1 a 1 caso existam valores negativos, funciona muito bem para trabalhar com dados dispersos / espalhados.



```python
# Selecionar apenas as colunas numéricas para normalização
colunas_numericas = dados.select_dtypes(include=[float, int]).columns

# Criar um objeto de normalização para dados dispersos
scaler = MaxAbsScaler()

# Normalizar os dados dispersos
dados[colunas_numericas] = scaler.fit_transform(dados[colunas_numericas])

dados
```

###Normalizando dados com outliers

Se os dados tiverem muitos outliers, utilizar os métodos mostrados anteriormente pode não funcionar tão bem, isso porque os outliers vão ter dados iguais a 1 e os demais dados vão ficar muito próximos de zero. Como alternativa podemos utilizar o RobustScaler

```python
from sklearn.preprocessing import RobustScaler
```

```python
# Selecionar apenas as colunas numéricas para normalização
colunas_numericas = dados.select_dtypes(include=[float, int]).columns

# Criar um objeto de normalização robusta
scaler = RobustScaler()

# Normalizar os dados com outliers
dados[colunas_numericas] = scaler.fit_transform(dados[colunas_numericas])

dados
```

O RobustScaler é uma técnica robusta, que utiliza a mediana e o intervalo interquartil (IQR) para dimensionar os dados, tornando-a menos sensível aos outliers. Assim, os valores são normalizados considerando a dispersão dos dados em torno da mediana, o que ajuda a evitar que os outliers distorçam a escala.

Essa abordagem é especialmente útil quando você tem dados que contenham muitos outliers ou quando a distribuição dos dados é altamente assimétrica.

##Discretização

A discretização é o processo de transformar valores contínuos em valores discretos ou categorias. Vamos prosseguir com a discretização dos dados do conjunto "Wine Quality" usando a biblioteca Scikit-learn e a técnica KBinsDiscretizer.

```python
from sklearn.preprocessing import KBinsDiscretizer

# Selecionar apenas as colunas numéricas para discretização
colunas_numericas = dados.select_dtypes(include=[float, int]).columns

# Definir o número de intervalos desejados
n_bins = 5

# Criar um objeto de discretização
discretizer = KBinsDiscretizer(n_bins=n_bins, encode='ordinal', strategy='uniform')

# Discretizar os dados
dados[colunas_numericas] = discretizer.fit_transform(dados[colunas_numericas])

dados
```

O KBinsDiscretizer divide os valores contínuos em "n_bins" intervalos discretos, onde "n_bins" é o número de intervalos que você deseja criar para cada coluna numérica.

O parâmetro encode='ordinal' especifica que as categorias serão codificadas como números inteiros, e o parâmetro strategy='uniform' indica que os intervalos serão criados com largura uniforme.

Depois de discretizar os dados, os valores em cada coluna numérica serão substituídos pelos intervalos discretos correspondentes.

##Binarização de variáveis

A binarização é o processo de transformar valores numéricos em valores binários (0 ou 1) com base em um limiar específico. Vamos prosseguir com a binarização das variáveis do conjunto "Wine Quality" usando a técnica Binarizer.

```python
from sklearn.preprocessing import Binarizer

# Selecionar apenas as colunas numéricas para binarização
colunas_numericas = dados.select_dtypes(include=[float, int]).columns

# Definir o limiar para a binarização
threshold = 5.0

# Criar um objeto de binarização
binarizer = Binarizer(threshold=threshold)

# Binarizar as variáveis
dados[colunas_numericas] = binarizer.fit_transform(dados[colunas_numericas])

dados.head(100)
```

O Binarizer transforma os valores em 0 ou 1 com base no limiar definido. Valores maiores que o limiar são transformados em 1, enquanto valores iguais ou menores que o limiar são transformados em 0.

No código acima, definimos o limiar como 5.0. Portanto, todos os valores maiores que 5.0 em cada coluna numérica serão binarizados para 1 e todos os valores iguais ou menores que 5.0 serão binarizados para 0.

A binarização é útil quando você deseja transformar valores contínuos em variáveis binárias ou quando deseja trabalhar com modelos que requerem entradas binárias.

Lembre-se de que a binarização pode levar à perda de informações, pois os valores contínuos são transformados em apenas duas categorias (0 ou 1). Portanto, é importante considerar cuidadosamente se a binarização é apropriada para o seu problema específico.

##Feature engineering (Criação de novas variáveis a partir das existentes)

A feature engineering envolve a criação de novas variáveis com base nas informações disponíveis no conjunto de dados existente. Vamos prosseguir com uma simples feature engineering no conjunto "Wine Quality", criando uma nova coluna que indica se o vinho é tinto ou branco com base na coluna "type".

```python
# Criar uma nova coluna "is_red" indicando se o vinho é tinto (1) ou branco (0)
dados['is_red'] = dados['type'].map({'red': 1, 'white': 0})

dados
```

A coluna "type" original contém os valores "red" e "white" para indicar se o vinho é tinto ou branco, respectivamente. Através da feature engineering, criamos uma nova coluna chamada "is_red" que terá o valor 1 para vinhos tintos e 0 para vinhos brancos.

Essa nova coluna "is_red" pode ser útil para a análise de dados e também pode ser usada como uma variável de entrada em modelos de aprendizado de máquina.

Além disso, a feature engineering pode envolver diversas outras transformações, como combinação de variáveis existentes, criação de variáveis categóricas a partir de variáveis contínuas, criação de variáveis polinomiais ou interações entre variáveis, entre outras técnicas.
